```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
res = pd.read_csv('shortest_path_neurips_exp.csv')
```

```python
agg_res = res.loc[res.groupby(['n','eps_type','trial','loss_name'])['val_metric'].idxmin()].groupby(['n','eps_type','loss_name']).agg({'test_regret': ['mean','std','count']}).reset_index()
```

```python
agg_res.columns = agg_res.columns.map('_'.join).str.strip('_')
```

```python
sns.lineplot(x='n', y='test_regret_mean', hue='loss_name', data=agg_res[agg_res.eps_type=='normal'], marker='o')
```

```python
sns.lineplot(x='n', y='test_regret_mean', hue='loss_name', data=agg_res[agg_res.eps_type=='unif'], marker='o')
```

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

group_A = agg_res[agg_res.eps_type=='normal']
group_B = agg_res[agg_res.eps_type=='unif']

# Left subplot: Plot for group A
# Lineplot with groups based on `marker_group_col`
for group, group_data in group_A.groupby('loss_name'):
    axes[0].errorbar(
        group_data['n'],
        group_data['test_regret_mean'],
        yerr=group_data['test_regret_std'],
        label=f'Group {group}',
        marker='o',
        linestyle='-'
    )
axes[0].set_title('normal')
axes[0].set_xlabel('n')
axes[0].set_ylabel('test_regret')
axes[0].legend()

# Right subplot: Plot for group B
for group, group_data in group_B.groupby('loss_name'):
    axes[1].errorbar(
        group_data['n'],
        group_data['test_regret_mean'],
        yerr=group_data['test_regret_std'],
        label=f'Group {group}',
        marker='o',
        linestyle='-'
    )
axes[1].set_title('unif')
axes[1].set_xlabel('test_regret_mean')
axes[1].legend()

# Adjust layout for clarity
plt.tight_layout()

plt.show()
```
